```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# Plot a scatter plot with age on the x-axis and charges on the y-axis to show how age affects insurance costs
x = insurance_data.loc[:,'age'].values
y = insurance_data.loc[:,'charges'].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatterplot of Age vs. Charges')
plt.grid(True)
plt.show()
```

```python
# Plot a scatter plot with age on the x-axis and charges on the y-axis to show how age affects insurance costs
# First we reshape the data to comply with Scikit-learn
x = insurance_data.loc[:,'age'].values.reshape(-1,1)
y = insurance_data.loc[:,'charges'].values.reshape(-1,1)

# Fit a model and make predictions on the data
insurance_model = LinearRegression()
insurance_model.fit(x,y)

# Create a prediction model and initialise unk_x as the unknown value.
unk_x = [[30]]
x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

# Predict unknown values for ages 30, 50 and 70
unk_x = [[30]]
print("When age is 30 charges are: ", int(insurance_model.predict(unk_x)))
unk_x = [[50]]
print("When age is 50 charges are: ", int(insurance_model.predict(unk_x)))
unk_x = [[70]]
print("When age is 70 charges are: ", int(insurance_model.predict(unk_x)))

# Plot another scatter plot with the line of best-fit.
plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatterplot of Age vs. Charges with line of best-fit')
plt.grid(True)
plt.show()
```
